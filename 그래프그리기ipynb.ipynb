```python
import matplotlib.pyplot as plt
import numpy as np


#    
years = np.array([2018, 2020, 2022, 2024, 2026, 2028, 2030])
market_sizes_A = np.array([70, 184.5, 326, 439.2, 657.3, 927.2, 1179.0])
market_sizes_B = np.array([2485.0,  2309.9, 2685.6, 2839.9, 3309.1, 3528.0, 3769.1])
percent = (market_sizes_A / market_sizes_B)*100


#  
plt.figure(figsize=(10, 10))  #   
bars_A = plt.bar(years, market_sizes_A, color='red',label='AI semiconductor',width=1)
bars_B = plt.bar(years, market_sizes_B, color='green', label='System semiconductor', alpha=0.3,width=1)
plt.xlabel('year')  # x  
plt.ylabel('market size(million $)')  # y  
plt.yticks(rotation=90)
plt.title('Semiconductor market trends ')  #   
plt.xticks(years)  # x   
plt.grid(True)  #  
plt.legend()

#      .
for bar, market_size, in zip(bars_B, market_sizes_B):
    plt.text(bar.get_x() + bar.get_width() /2, bar.get_height() + 100, str(market_size), ha='center', color='black')
for bar, market_size,per in zip(bars_A, market_sizes_A, percent.round(1)):
    plt.text(bar.get_x() + bar.get_width() / 2 , bar.get_height() + 200, str(market_size), ha='center', color='red')
    plt.text(bar.get_x() + bar.get_width() / 2 , bar.get_height() + 100, f'({per:.1f}%)', ha='center', color='red')


plt.show()  #  
```
